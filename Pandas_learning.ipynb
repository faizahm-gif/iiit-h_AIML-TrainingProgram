```python
#apply(series and dataframe): to perform calculations on columns
points=[[(1,2),(2,3)],[(1,2),(3,3)],[(1,2),(4,3)]]
pts=pd.DataFrame(points,columns=['Point A', 'Point B'])
pts
```

```python
def dist(pts):
  A=pts['Point A']
  B=pts['Point B']
  return ((A[0]-B[0])**2+(A[1]-B[1])**2)**0.5
```

```python
pts['Distance']=pts.apply(dist,axis=1)
```
