```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h1>Subplots
<p> To plot multiple graphs side by side


```python
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batsman['avg'],batsman['strike_rate'])
ax.set_xlabel('average')
ax.set_ylabel('Strike Rate')
ax.set_title('Batsman Data')
fig.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(15,6))
ax[0].scatter(batsman['avg'],batsman['runs'])
ax[0].set_xlabel('average')
ax[0].set_ylabel('Runs')
ax[0].set_title('Batsman Data')
ax[1].scatter(batsman['avg'],batsman['strike_rate'])
ax[1].set_xlabel('average')
ax[1].set_ylabel('Strike Rate')
ax[1].set_title('Batsman Data')
fig.show()
# for 2 col and 2 rows we use [][] matrix indexing such as ax[1][1], ax[1][0]
```

```python
fig = plt.figure()
ax= plt.subplot(projection='3d')
ax.scatter3D(batsman['avg'],batsman['runs'],batsman['strike_rate'],marker='+')
ax.plot3D(batsman['avg'],batsman['runs'],batsman['strike_rate'],color='red')
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
```

```python
z =x**2+yy**2
z.shape
```

```python
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis',)
fig.colorbar(p)
```

```python
z= np.sin(xx)+np.cos(yy)
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis',)
fig.colorbar(p)
```

```python
#CONTOUR PLOT : it is a 2D graph of a 3D graph
z= np.sin(xx)+np.cos(yy)
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis',)
fig.colorbar(p)
```

```python
z= np.sin(xx)+np.cos(yy)
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis',)
fig.colorbar(p)
```
